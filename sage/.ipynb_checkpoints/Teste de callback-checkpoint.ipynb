```python
def default_arranging(n,k):
    p = MixedIntegerLinearProgram(maximization=True,solver=my_solver)

    G=graphs.CompleteGraph(n+1)
    for e in G.edges():
        if e[1]-e[0] > k:
            G.delete_edge(e)
    
#    variável da aresta: x[e]=1 implica que e está no nosso caminho
    x = p.new_variable(binary=True)

#    equações dos vértices
    for u in G.vertices():
        equation=0
        for e in G.edges_incident(u):
            equation+=x[e]
        if u==0:
            p.add_constraint(equation==1)
        else:
            p.add_constraint(equation<=2)
    
    for i in range(1,k+1):
        equation=0
        for e in G.edges():
            if e[1]-e[0] == i:
                equation+=x[e]
        if i>=2*k-n+1:
            p.add_constraint(equation==2)
        else:
            p.add_constraint(equation==1)
    return p,x,G
```

```python
def arranging_maximizing_small_cuts(n,k,should_print):

    p,x,G=default_arranging(n,k)

# Daqui pra frente definimos o objetivo
    objective=0
#     y=p.new_variable(binary=True)
#     for i in range(n-2):
#         for j in range(i+2,n):
#             H=G.subgraph([i,i+1,j,j+1])
#             equation=0
#             for e in H.edges():
#                 equation+=x[e]
#             p.add_constraint(y[i,i+1,j,j+1] <= (1/3)*equation)
#             objective+=y[i,i+1,j,j+1]
#     p.set_objective(objective)
    
    y=p.new_variable(binary=True)
    for i in range(n-2):
        j=n-i-1
        H=G.subgraph(list(range(i))+list(range(j,n+1)))
        equation=0
        for e in H.edges():
            equation+=x[e]
        p.add_constraint(y[i] <= (1/(H.order()-1))*equation)
        objective+=y[i]
    p.set_objective(objective)
    
    sol=p.solve()
    solution=p.get_values(x).items()
    R=[]
    for s in solution:
        if s[1]==1:
            R.append(s[0])
    H=Graph(R)

    while H.is_tree()==False:
        comps=H.connected_components()
        for comp in comps:
            X=H.subgraph(comp)
            if X.size() > X.order()-1:
                equation=0
                for e in X.edges():
                    equation+=x[e]
                p.add_constraint(equation<=X.order()-1)
        sol=p.solve()
        solution=p.get_values(x).items()
        R=[]
        for s in solution:
            if s[1]==1:
                R.append(s[0])
        H=Graph(R)

    if should_print:    
        H.show(layout="circular")
        print("number of parallel pairs: "+str(sol))
    lista=[0]
    u=0
    while H.order()>=2:
        v=H.neighbors(u)[0]
        H.delete_vertex(u)
        lista.append(v)
        u=v
    return lista            
```

```python
#!/bin/env sage
# -*- coding: utf-8 -*-
import gurobipy as gp
from gurobipy import GRB

def mycallback(model, where):
    """
    This function checks whether the incumbent solution found by Gurobi has a
    cycle or not. If a cycle was found, then this function adds a constraint
    that is violated by the current solution.

    To be more precise, if a cycle C is found in the current solution,
    then this function adds the following constraint

      ∑ x_{e, 0} ≤ |E(C)| - 1
    e ∈ E(C)
    """

    if where == GRB.Callback.MIPSOL:

        vals = model.cbGetSolution(model._x)

        H = Graph()
        for (e, val) in vals.items():
            if val > 0:
                u,v,label=e
                H.add_edge(u, v)

        is_forest, cycle = H.is_forest(certificate=True)
        if not is_forest:
            csize = len(cycle)
            cycle.append(cycle[0])
            expr = 0
            ids = []
            for i in range(csize):
                u=cycle[i]
                v=cycle[i+1]
                if u < v:
                    expr += model._x[(u,v,None)]
                else:
                    expr += model._x[(v,u,None)]
#            print("restriction added on cycle",cycle)

            model.cbLazy(expr <= csize - 1)
        

class Model:

    def __init__(self, n,k):

        self.model = gp.Model("arranging_with_restrictions")
        self.model.setParam('OutputFlag', 0)
        self.n = n
        self.k = k
        
        G = graphs.CompleteGraph(n+1)
        for e in G.edges():
            if e[1]-e[0] > k:
                G.delete_edge(e)
            self.G = G

        self._init_x_variables()

        # The next lines are defining some variables inside the model object so
        # that they can be accessed in the callback (yeah.. I'm breaking
        # encapsulation).  That's not very pretty, but it's the way it's
        # suggested by the Gurubi Another solution would be to define global
        # variables.
        self.model._x = self.x
        self.model._G = self.G

        self._add_constr_vertex_degree()
        self._add_constr_which_edge_repeat()


    def _init_x_variables(self):
        """
              ⎧1,  if e ∈ E(G) is selected
        x_e = ⎨
              ⎩0,  otherwise
        """
        indices = [e for e in self.G.edges()]
        self.x = self.model.addVars(indices, lb=0.0, ub=1.0, vtype=GRB.BINARY, name="x")



    def _add_constr_vertex_degree(self):
        """
        Every vertex has at degree at most 2, and 0 has degree precisely 1

        """

        for u in self.G.vertices():
            equation=0
            for e in self.G.edges_incident(u):
                equation+=self.x[e]
            if u==0:
                self.model.addConstr(equation==1, name='starts at 0')
            else:
                self.model.addConstr(equation<=2, name='max degree 2')
                

    def _add_constr_which_edge_repeat(self):
        """
        Some edge have to repeat once, and some edge do not repeat

        """
        
        for i in range(1,self.k+1):
            equation=0
            edges=[]
            for e in self.G.edges():
                if e[1]-e[0] == i:
                    edges.append((e[0],e[1]))
                    equation+=self.x[e]
            if i>=2*self.k-self.n+1:
                self.model.addConstr(equation==2, name='this edge repeats')
#                print("restrição do peso",i,"igual a 2")
#                print(edges)
            else:
                self.model.addConstr(equation==1, name='this edge does not repeat')
#                 print("restrição do peso",i,"igual a 1")
#                 print(edges)

    def solve(self):

        self.model.write("modelo_debug.lp")

        # ⚠ if you plan to use lazyConstraints in Gurobi,
        #     you must set the following parameter to 1
        self.model.Params.lazyConstraints = 1

        # to use an lazyConstraint `foo`, you must pass it
        # as a parameter in the function optimize
        self.model.optimize(mycallback)

    def path_edges(self):

        if self.model.status != GRB.OPTIMAL:
            return []

        edges=[]
        for e in self.G.edges():
#             print(self.x[e].x)
            if self.x[e].x >0.1:
                edges.append(e)
        return edges


def show_path(edges):
    
    H=Graph(edges)
    lista=[0]
    u=0
    while H.order()>=2:
        v=H.neighbors(u)[0]
        H.delete_vertex(u)
        lista.append(v)
        u=v
    return lista
```
